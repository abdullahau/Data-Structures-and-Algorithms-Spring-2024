## Solution

When the robot moves, we keep track of the visited squares in the set `visited`. The elements of the set are pairs representing the coordinates of the squares.

```python
def count(s):
    visited = set()
    visited.add((0,0))
    y = x = 0
    for c in s:
        if c == "U": y -= 1
        if c == "D": y += 1
        if c == "L": x -= 1
        if c == "R": x += 1
        visited.add((y,x))
    return len(visited)
```
