# Binary Tree

## Sum of all Nodes

```python
# Binary Tree - Left & Right Recursion Method

class Node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def sum(tree):
    if tree.left == None and tree.right == None:
        return tree.value
    else:
        leftTree = sum(tree.left)
        rightTree = sum(tree.right)
        return tree.value + leftTree + rightTree

tree = Node(5, 
            Node(11, 
                  Node(4, 
                        Node(14),
                         Node(3)),
                   Node(8, 
                        Node(10),
                         Node(2))),
             Node(3, 
                  Node(9, 
                        Node(6),
                         Node(15)),
                   Node(12, 
                        Node(1),
                         Node(7))))

sum(tree)
```

## Max Value of Node

```python
# Binary Tree - Left & Right Recursion Method

class Node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def maxTree(tree):
    if tree.left == None and tree.right == None:
        return tree.value
    else:
        leftTree = maxTree(tree.left)
        rightTree = maxTree(tree.right)
        return max(tree.value, leftTree, rightTree)

tree = Node(5, 
            Node(11, 
                  Node(4, 
                        Node(14),
                         Node(3)),
                   Node(8, 
                        Node(10),
                         Node(2))),
             Node(3, 
                  Node(9, 
                        Node(6),
                         Node(15)),
                   Node(12, 
                        Node(1),
                         Node(7))))

maxTree(tree)
```

## Tree Height

```python
# Binary Tree - Left & Right Recursion Method

class Node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def heightTree(tree):
    if tree.left == None and tree.right == None:
        return 0
    else:
        leftTree = heightTree(tree.left)
        rightTree = heightTree(tree.right)
        return 1 + max(leftTree, rightTree)

tree = Node(5, 
            Node(11, 
                  Node(4, 
                        Node(14),
                         Node(3)),
                   Node(8, 
                        Node(10),
                         Node(2))),
             Node(3, 
                  Node(9, 
                        Node(6),
                         Node(15)),
                   Node(12, 
                        Node(1),
                         Node(7))))

heightTree(tree)
```

## Exists in Tree

```python
# Binary Tree - Left & Right Recursion Method

class Node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def searchTree(tree, num):
    if tree.left is None:
        return tree.value == num
    else:
        leftTree = searchTree(tree.left, num)
        rightTree = searchTree(tree.right, num)
        return tree.value == num or leftTree or rightTree

tree = Node(5, 
            Node(11, 
                  Node(4, 
                        Node(14),
                         Node(3)),
                   Node(8, 
                        Node(10),
                         Node(2))),
             Node(3, 
                  Node(9, 
                        Node(6),
                         Node(15)),
                   Node(12, 
                        Node(1),
                         Node(7))))

searchTree(tree, 14)
```

## Reverse Binary Tree

```python
# Binary Tree - Left & Right Recursion Method

class Node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
        
    def __repr__(self) -> str:
        if self.left is not None:
            return f"Node({self.value}, {self.left}, {self.right})"
        else:
            return f"Node({self.value})"

def reverseTree(tree):
    if tree.left is None:
        return
    else:
        reverseTree(tree.left)
        reverseTree(tree.right)
        tree.left, tree.right = tree.right, tree.left

tree = Node(5, 
            Node(11, 
                  Node(4, 
                        Node(14),
                         Node(3)),
                   Node(8, 
                        Node(10),
                         Node(2))),
             Node(3, 
                  Node(9, 
                        Node(6),
                         Node(15)),
                   Node(12, 
                        Node(1),
                         Node(7))))

reverseTree(tree)
print(tree)

reversed = Node(5, 
                Node(3, 
                     Node(12, 
                          Node(7), 
                          Node(1)), 
                     Node(9, 
                          Node(15), 
                          Node(6))), 
                Node(11, 
                     Node(8, 
                          Node(2), 
                          Node(10)), 
                     Node(4, 
                          Node(3), 
                          Node(14))))
```
